A notebook to demo various functionalities and recipes.

# Graph the attribute dependencies of an object

```python
from i2 import Pipe
from i2.footprints import attribute_dependencies
from meshed.itools import graphviz_digraph

attribute_dependency_graph = Pipe(attribute_dependencies, dict, graphviz_digraph)
```

```python
# Defines classes for implementing web servers.
from http.server import HTTPServer
attribute_dependency_graph(HTTPServer)
```

```python
# Represents an entire XML document as a tree and provides a simple and efficient API for parsing and creating XML data.
from json import JSONDecoder
attribute_dependency_graph(JSONDecoder)
```

```python
# ElementTree is a simple and lightweight XML processor that provides an easy way 
# to access and manipulate XML documents.
from xml.etree.ElementTree import ElementTree
attribute_dependency_graph(ElementTree)
```

```python
# Implements a decoder for JSON, which can parse JSON from strings or files.
from logging import Logger
attribute_dependency_graph(Logger)
```

```python
# Provides access to the BSD socket interface.
from pathlib import Path
attribute_dependency_graph(Path)
```

```python
# Supplies classes for manipulating dates and times.
from threading import Thread
attribute_dependency_graph(Thread)
```

```python
# Represents an activity that is run in a separate process.
from socket import socket
attribute_dependency_graph(socket)
```
